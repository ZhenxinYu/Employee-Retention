```python
# Create the dataframe to store data
reviewsDF = pd.DataFrame(columns=['title', 'href', 'time', 'rating', 'authorJobTitle', 'authorLocation', 'recommends', 'mainText', 'pros', 'cons'])
url = 'https://www.glassdoor.com/Reviews/Google-Reviews-E9079.htm'
```

```python
while True:
    try:
        browser.get(url)
        for li in browser.find_elements_by_xpath('//li[@class="empReview cf"]|//li[@class=" empReview cf "]'):
            data = {}

            data['title'] = li.find_element_by_xpath('.//a[@class="reviewLink"]').text
            data['href'] = li.find_element_by_xpath('.//a[@class="reviewLink"]').get_attribute("href")
            data['time'] = li.find_element_by_xpath('.//time').get_attribute("datetime")
            data['rating'] = li.find_element_by_xpath('.//span[@class="rating"]/span').get_attribute("title")

            title = li.find_elements_by_xpath('.//span[starts-with(@class, "authorJobTitle")]')
            data['authorJobTitle'] = None if not title else title[0].text
            loc = li.find_elements_by_xpath('.//span[starts-with(@class, "authorLocation")]')
            data['authorLocation'] = None if not loc else loc[0].text

            recommends = li.find_elements_by_xpath('.//div[@class="flex-grid recommends"]//span')
            recommends += li.find_elements_by_xpath('.//div[@class="row reviewBodyCell recommends"]//span')
            data['recommends'] = [i.text for i in recommends]
            try:
                data['mainText'] = li.find_element_by_xpath('.//p[starts-with(@class, "mainText")]|.//p[@class=" tightBot mainText"]').text
            except:
                data['mainText'] = None

            data['pros'] = li.find_element_by_xpath('.//p[starts-with(., "Pros")]/../p[2]').text
            data['cons'] = li.find_element_by_xpath('.//p[starts-with(., "Cons")]/../p[2]').text

            reviewsDF.loc[len(reviewsDF)] = data
        
    except Exception as e:
        print(e)
        time.sleep(30)
        continue
    try:
        nextPage = browser.find_element_by_xpath('//a/span[text()="Next"]/..|//a/i/span[text()="Next"]/../..')
        url = nextPage.get_attribute("href")
    except:
        break
    reviewsDF.to_csv('reviews.csv')
```
