# Pros

## (a) Frequency

```python
# Get unigram, bigram, trigram words 
stopwords = stopwords.union({'worklife', 'work', 'life', 'balance', 'â€¦'})
bagOfWords = Counter()
for index, row in software.iterrows():
    pros = row['pros'].lower().translate(str.maketrans('', '', string.punctuation))
    unigram = [w for w in pros.split() if w not in stopwords]
    bigram = ['_'.join(i) for i in nltk.bigrams(nltk.word_tokenize(pros)) if i[0] not in stopwords and i[1] not in stopwords]
    trigram = ['_'.join(i) for i in nltk.trigrams(nltk.word_tokenize(pros)) if i[0] not in stopwords and i[1] not in stopwords and i[2] not in stopwords]
    bagOfWords.update(unigram)
    bagOfWords.update(bigram*4)
    bagOfWords.update(trigram*9)
```

```python
# Compute the word cloud
def red_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl({}, 100%, {}%)".format(random.choice([125,236]), random.randint(50,90))
url = 'http://i.dailymail.co.uk/i/pix/2015/09/01/18/2BE1E88B00000578-3218613-image-m-5_1441127035222.jpg'
img = np.array(Image.open(BytesIO(requests.get(url).content)))
wordcloud = WordCloud(background_color="white", mask=img)
wordcloud.generate_from_frequencies(frequencies=bagOfWords)
wordcloud.recolor(color_func = red_color_func)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## (b) Inverse Document Fequency

```python
stopwords = set(nltk.corpus.stopwords.words('english'))
bagOfWords = Counter()
for index, row in reviews.iterrows():
    pros = row['pros'].lower().translate(str.maketrans('', '', string.punctuation))
    bagOfWords.update(set(pros.split()))
for word in stopwords:
    if word in bagOfWords:
        del bagOfWords[word]
```

```python
for index, row in reviews.iterrows():
    pros = row['pros'].lower().translate(str.maketrans('', '', string.punctuation))
    bigrams = ['_'.join(i) for i in nltk.bigrams(nltk.word_tokenize(pros)) if i[0] not in stopwords and i[1] not in stopwords]
    bagOfWords.update(set(bigrams))
```

```python
for index, row in reviews.iterrows():
    pros = row['pros'].lower().translate(str.maketrans('', '', string.punctuation))
    trigrams = ['_'.join(i) for i in nltk.trigrams(nltk.word_tokenize(pros)) if i[0] not in stopwords and i[1] not in stopwords and i[2] not in stopwords]
    bagOfWords.update(set(trigrams))
```

```python
for w in bagOfWords:
    bagOfWords[w] = -np.log(len(reviews)/bagOfWords[w])
```

# Model

```python
# Prepare the data
stopwords = set(nltk.corpus.stopwords.words('english'))
stopwords.add('â€¦')
stopwords.add('within')
def sentence_preprocessing(sentence):
    sentence = sentence.lower()
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    words = sentence.split()
    return [w for w in words if w not in stopwords]

filters = {'feel_like', 'working_google', 'like_youre', 'work_really', 'work_google', 'work_may',\
           'wont_get', 'easy_feel', 'hard_make', 'might_feel', 'work_that', 'ive_seen', 'easy_get',\
           'like_big', 'get_work', 'dont_think', 'google_become', 'may_take', 'software_engineer',\
           'software_engineering', 'ive_worked', 'anywhere_else', 'dont_like', 'take_lots', 'also_many'}

def bigram(words):
    return ['_'.join(i) for i in nltk.bigrams(nltk.word_tokenize(' '.join(words))) if '_'.join(i) not in filters]

def trigram(words):
    return ['_'.join(i) for i in nltk.trigrams(nltk.word_tokenize(' '.join(words)))]

data = software
```

```python
# Count the words
idf_pros = Counter()
for index, row in data.iterrows():
    pros = sentence_preprocessing(row['pros'])
    #idf_pros.update(set(pros))
    idf_pros.update(set(bigram(pros)))
    idf_pros.update(set(trigram(pros)))
for w in idf_pros:
    idf_pros[w] = np.log(len(reviews)/idf_pros[w])
```

```python
# Count the words
idf_cons = Counter()
for index, row in data.iterrows():
    cons = sentence_preprocessing(row['cons'])
    #idf_cons.update(set(cons))
    idf_cons.update(set(bigram(cons)))
    idf_cons.update(set(trigram(cons)))
for w in idf_cons:
    idf_cons[w] = np.log(len(reviews)/idf_cons[w])
```

```python
# Split into the train and test sets
train = data.sample(frac=0.9)
test = data[~data.index.isin(train.index)]
```

```python
# Regression Model
lam, lr = 7, 0.005
prev_MSE, MSE = 10000, 9999
trainCountPros, trainCountCons = Counter(), Counter()
for idx, row in train.iterrows():
    pros_word, cons_word = sentence_preprocessing(row['pros']), sentence_preprocessing(row['cons'])
    trainCountPros.update(set({}).union(set(bigram(pros_word))).union(set(trigram(pros_word))))
    trainCountCons.update(set({}).union(set(bigram(cons_word))).union(set(trigram(cons_word))))
alpha = random.random()
beta_pros = {i:random.random() for i in trainCountPros}
beta_cons = {i:random.random() for i in trainCountCons}
while prev_MSE-MSE > 10**(-5):
    prev_MSE = MSE
    dalpha = 0
    dbeta_pros = {i:2*lam*beta_pros[i] for i in trainCountPros}
    dbeta_cons = {i:2*lam*beta_cons[i] for i in trainCountCons}
    for idx, row in train.iterrows():
        pros_word, cons_word = sentence_preprocessing(row['pros']), sentence_preprocessing(row['cons'])
        pros_counter = Counter(bigram(pros_word)+trigram(pros_word))
        cons_counter = Counter(bigram(cons_word)+trigram(cons_word))
        tfidf_pros = {i:pros_counter[i]/len(pros_word)/2*idf_pros[i] for i in pros_counter}
        tfidf_cons = {i:cons_counter[i]/len(cons_word)/2*idf_cons[i] for i in cons_counter}
        prediction = alpha + sum([tfidf_pros[i]*beta_pros[i]*len(i.split('_')) for i in tfidf_pros]) + sum([tfidf_cons[i]*beta_cons[i]*len(i.split('_')) for i in tfidf_cons])
        err = prediction - (row['Status']=='Former Employee ')
        dalpha += 2*err
        for i in pros_counter:
            dbeta_pros[i] += 2*err
        for i in cons_counter:
            dbeta_cons[i] += 2*err
    alpha -= lr*dalpha/len(train)
    beta_pros = {i:beta_pros[i]-lr*dbeta_pros[i]/trainCountPros[i] for i in trainCountPros}
    beta_cons = {i:beta_cons[i]-lr*dbeta_cons[i]/trainCountCons[i] for i in trainCountCons}
    MSE_l = []
    for idx, row in test.iterrows():
        try:
            pros_word, cons_word = sentence_preprocessing(row['pros']), sentence_preprocessing(row['cons'])
            pros_counter = Counter(bigram(pros_word)+trigram(pros_word))
            cons_counter = Counter(bigram(cons_word)+trigram(cons_word))
            tfidf_pros = {i:pros_counter[i]/len(pros_word)/2*idf_pros[i] for i in pros_counter}
            tfidf_cons = {i:cons_counter[i]/len(cons_word)/2*idf_cons[i] for i in cons_counter}
            prediction = alpha + sum([tfidf_pros[i]*beta_pros[i]*len(i.split('_')) for i in tfidf_pros]) + sum([tfidf_cons[i]*beta_cons[i]*len(i.split('_')) for i in tfidf_cons])
            MSE_l.append(((row['Status']=='Former Employee ')-prediction)**2)
        except KeyError:
            pass
    MSE = np.mean(MSE_l)
    print('Validation MSE: {}, {}'.format(MSE, len(MSE_l)))
```

```python
plt.figure(figsize=(20,5))
#plt.subplot(2,1,1)
pros = sorted(beta_pros.items(), key=lambda x:x[1], reverse=True)[:10]
pros2 = sorted(beta_pros.items(), key=lambda x:x[1], reverse=True)[-10:][::-1]
cons = sorted(beta_cons.items(), key=lambda x:x[1], reverse=True)[:10][::-1]
cons2 = sorted(beta_cons.items(), key=lambda x:x[1], reverse=True)[-10:]
plt.bar(*zip(*(pros+cons)), color=['b']*10+['r']*10)
plt.xticks(rotation=35)
plt.title('Current Employees')
plt.show()
plt.figure(figsize=(20,5))
#plt.subplot(2,1,2)
plt.bar(*zip(*(pros2+cons2)), color=['b']*10+['r']*10)
plt.xticks(rotation=35)
plt.title('Former Employees')
plt.show()
```
